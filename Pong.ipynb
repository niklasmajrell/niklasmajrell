## Pong 
Pong is a two-dimensional sports game that simulates table tennis. The player controls an in-game paddle by moving it vertically across the left or right side of the screen. They can compete against another player controlling a second paddle on the opposing side. Players use the paddles to hit a ball back and forth. The goal is for each player to reach eleven points before the opponent; points are earned when one fails to return the ball to the other. 

```python
import turtle as t
import time

# Creating objects: 
# Screen
screen = t.Screen()
screen.title("NOD Pong Cup")
screen.bgcolor("black")
screen.setup(width=1000, height=600)

# Paddles 
left_pad = t.Turtle() 
left_pad.shape("square")
left_pad.color("white")
left_pad.shapesize(stretch_wid=6, stretch_len=2) 

left_pad.penup() 
left_pad.goto(-400, 0) 
left_pad.speed(0) 

right_pad = t.Turtle()
right_pad.shape("square")
right_pad.color("white")
right_pad.shapesize(stretch_wid=6, stretch_len=2) 

right_pad.penup() 
right_pad.goto(400, 0) 
right_pad.speed(0) 

# Scoreboard 
pen = t.Turtle()
pen.speed(0)
pen.color('hotpink')
pen.penup()
pen.hideturtle()
pen.goto(0,260)
pen.write("Player 1: 0                    Player 2: 0 ",align="center",font=('Courier',24,"normal"))

# Ball
ball = t.Turtle()
ball.speed(0)
ball.shape("circle") 
ball.color("yellow")
ball.penup()
ball.goto(0, 0)
ball.shapesize(2)
ball.dx = 5
ball.dy = -5

# Objects end

# Score variables
player1_score = 0
player2_score = 0

rounds = t.numinput("title", "How many rounds do you want to play?")

def left_paddleup():
    '''
    Makes left paddle move upwards, no arg
    '''
    y = left_pad.ycor() 
    y += 50
    left_pad.sety(y) 

def left_paddledown():
    '''
    Makes left paddle move downwards, no arg
    '''
    y = left_pad.ycor()
    y += -50
    left_pad.sety(y)
    
def right_paddleup():
    '''
    Makes right paddle move upwards, no arg
    '''
    y = right_pad.ycor()
    y += 50
    right_pad.sety(y)

def right_paddledown():
    '''
    Makes right paddle move downwards, no arg
    '''
    y = right_pad.ycor()  
    y += -50
    right_pad.sety(y)

# Keyboard bindings for paddles
screen.onkeypress(left_paddleup, "q")
screen.onkeypress(left_paddledown, "a")
screen.onkeypress(right_paddleup, "Up")
screen.onkeypress(right_paddledown, "Down")

screen.listen()

# Main gameloop
while rounds > 0:
    screen.update()
    ball.setx(ball.xcor()+ball.dx)
    ball.sety(ball.ycor()+ball.dy)
    if left_pad.ycor() >= 260:
        left_paddledown()
    if left_pad.ycor() <= -260:
        left_paddleup()
    if right_pad.ycor() >= 260:
        right_paddledown()
    if right_pad.ycor() <= -260:
        right_paddleup()

    # Ball bounce
    if ball.ycor() < -300 or ball.ycor() > 300:
        ball.dy *= -1
            #     
    if (ball.xcor() > 375 and ball.xcor() < 400) and (ball.ycor() < right_pad.ycor() + 40 and ball.ycor() > right_pad.ycor() - 40):
        ball.dx *= -1
        ball.color("blue")
    if (ball.xcor() < -375 and ball.xcor() > -400) and (ball.ycor() < left_pad.ycor() + 40 and ball.ycor() > left_pad.ycor() - 40):
        ball.dx *= -1
        ball.color("green")
        
    # Counting score
    if ball.xcor() > 500:
            
        player1_score = player1_score + 1
        pen.clear()
        pen.write("Player 1: {}                    Player 2: {} ".format(player1_score,player2_score),align="center",font=('Monaco',24,"normal"))
        time.sleep(2)
        ball.goto(0, 0)
 
        rounds -= 1

    if ball.xcor() < -500:
        player2_score = player2_score + 1
        pen.clear()
        pen.write("Player 1: {}                    Player 2: {} ".format(player1_score,player2_score),align="center",font=('Monaco',24,"normal"))
        time.sleep(2)
        ball.goto(0, 0)
            
        rounds -= 1
    
    # Write out the winner on turtle screen 
    if rounds == 0 and player1_score > player2_score:
        pen.goto(0,0)
        pen.write("Player 1 wins!", align="center", font=('Courier',80,"normal"))

    if rounds == 0 and player2_score > player1_score:
        pen.goto(0,0)
        pen.write("Player 2 wins!", align="center", font=('Courier',80,"normal"))
    
    if rounds == 0 and player2_score == player1_score:
        pen.goto(0,0)
        pen.write("It's a draw!", align="center", font=('Courier',80,"normal"))

screen.mainloop() 
```
